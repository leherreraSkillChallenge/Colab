> **Estadística**

> **Estadística Descriptiva**


> Es el subcampo de la Estadística que se encarga de describir y resumir las características esenciales de un conjunto de datos mediante estadísticos como media, moda, mediana, desviación estándar, entre otros.

> **Estadística Prescriptiva**


> Es una de las ramas de la estadística que se ocupa de diseñar y analizar sistemas para tomar decisiones basadas en datos. Ella combina modelos estadísticos con conocimiento experto para generar recomendaciones o acciones óptimas en determinados contextos.










**3. Análisis Exploratorio de Datos**



> Análisis exploratorio de datos por sus siglas (EDA), se utiliza para analizar e investigar conjuntos de datos y resumir sus características principales, a menudo empleando métodos de visualización de datos. Ayuda a determinar la mejor manera de gestionar las fuentes de datos para obtener las respuestas que se  necesitan contestar, con lo cual se facilita el descubrimiento de patrones, anomalías y la conprobación de hipótesis o suposiciones.


> EDA se utiliza principalmente para ver qué pueden revelar los datos más allá del modelado formal o tarea de prueba de hipótesis y proporciona una mejor comprensión de las variables del conjunto de datos y las relaciones entre ellas.







> **3.1. Tipos de (EDA)**


> **Univariante no gráfico.**

> Los datos constan de una sola variable.


> **Univariante gráfico.**

> Los datos requieren apoyo mediante graficos, que pueden ser: Histogramas, Diagramas de Frecuencia, entre otros.


> **No gráfico multivariante.**

> Los datos surgen con el analisis de más de una variable.


> **Gráfico multivariante.**

> Datos se derivan de más de una variable y es necesario utilizar gráficos para mostrar las relaciones entre dos o más conjuntos de datos.

```python
#salario miminimo en Colombia 2024
minimo = 1300000
```

**5.3. Generar preguntas de análisis sobre los datos**

**5.3.1 Preguntas a Responder:**

> ¿Cual población es la población con mayor cantidad de pensionados?

> ¿Cual es el rango de edad de la población de pensionados?

**5.3.2 Normalización de los datos existentes**


**>Para responder a la pregunta pasaremos una variable categoria a numerica, aplicando los valores del salario mínimo a sus categorias**

```python
# Convertir la columna 'Cantidad Personas' a int, reemplazando las categorias por numeros
df['Rango salarial (SMLV)'] = df['Rango salarial (SMLV)'].replace({
    'Hasta 2': minimo*2,
    'Entre 2 y 4': minimo*4,
    'Entre 4 y 7': minimo*7,
    'Entre 7 y 10': minimo*10,
    'Entre 10 y 13': minimo*13,
    'Entre 13 y 16': minimo*16,
    'Mayor a 16': minimo*17
}).astype(int)
```

```python
# Ver las primeras filas de los datos
print(df.head())
```

**>Renombrar variables existentes para mejorar su interpretación a la leer los datos**

```python
# Convertir la columna 'Rango salarial (SMLV)' a 'Rango Pensional (SMLV)'
df['Rango Pensional (SMLV)'] = df['Rango salarial (SMLV)']
df.drop('Rango salarial (SMLV)', axis=1, inplace=True)
```

```python
# Ver las primeras filas de los datos
print(df.head())
```

**>Hacer limpieza de datos, para poder utilizar los valores númericos en las mismas**

```python
# Convertir la columna 'Cantidad Personas' a int, reemplazando las comas
df['Cantidad Personas'] = df['Cantidad Personas'].str.replace(',', '').astype(int)
```

```python
# Ver las primeras filas de los datos
print(df.head())
```

```python
# Obtener información sobre los tipos de datos y los valores perdidos
print(df.info())
```

```python
# Describir estadísticas básicas de las columnas numéricas
print(df.describe())
```

**5.4. Encontrar en las visualizaciones respuestas a esas preguntas. Graficar todo lo que sea necesario**

Utilización de las librerias de matplotlib para visualización de datos

```python
# Crear un gráfico de barras para la columna 'Genero'
df['Genero'].value_counts().plot(kind='pie')
plt.show()
```

>**Grafica 5.4.1**

```python
# Crear un gráfico de barras para la columna 'Tipo Pension'
df['Tipo Pension'].value_counts().plot(kind='bar')
plt.show()
```

>**Grafica 5.4.2**

```python
# Crear un gráfico de distribución de frecuencias para la columna 'Rango Pensional (SMLV)' vs 'Genero'
sns.histplot(data=df, x='Rango Pensional (SMLV)', hue='Genero', multiple='stack')
plt.show()
```

>**Grafica 5.4.3**

```python
# Crear un gráfico de barras para la columna 'Genero' y 'Rango Edad' vs 'Cantidad Personas'
plt.figure(figsize=(10, 8))
sns.barplot(x='Genero', y='Cantidad Personas', hue='Rango Edad', data=df)
plt.show()
```

>**Grafica 5.4.4**

```python
# Crear un gráfico de distribución de frecuencias para la columna 'Cantidad Personas' vs 'Rango Edad'
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Cantidad Personas', y='Rango Edad')
plt.xlabel('Cantidad Personas')
plt.ylabel('Rango Edad')
plt.title('Boxplot de Cantidad de Personas por Rango de Edad')
plt.show()
```
